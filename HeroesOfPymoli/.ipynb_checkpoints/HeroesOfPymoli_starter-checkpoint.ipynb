```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
h_o_p = "Data/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_pd = pd.read_csv(h_o_p)
purchase_data_pd.head()
```

## Player Count

* Display the total number of players


```python
players = purchase_data_pd.loc[:,["SN", "Gender", "Age"]]
unique_players = players.drop_duplicates()
total_players = unique_players.count()[0]
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
average_price = purchase_data_pd["Price"].mean()
number_purchase = len(purchase_data_pd)
total_revenue = purchase_data_pd["Price"].sum()
unique_items = purchase_data_pd["Item ID"].unique()
total_items = len(unique_items)

purchase_summary_df = pd.DataFrame({"Number of Unique Items": [total_items],
                                    "Average Price": [average_price],
                                    "Number of Purchases": [number_purchase],
                                    "Total Revenue": [total_revenue]})

purchase_summary_df["Average Price"] = purchase_summary_df["Average Price"].map("${0:,.2f}".format)
purchase_summary_df["Total Revenue"] = purchase_summary_df["Total Revenue"].map("${0:,.0f}".format)

purchase_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
total_gender = unique_players["Gender"].value_counts()
total_gender_percent = total_gender / total_players

gender_demo_df = pd.DataFrame({"Total": total_gender, "Percent": total_gender_percent})
gender_demo_df["Percent"] = gender_demo_df["Percent"].map("{0:,.2%}".format)

gender_demo_df
#print(f" Number and Percentage of Male players: {male}, {male_percent}%\n Female players: {female}, {female_percent}%\n Other/Non-Disclosed players: {other}, {other_percent}%")
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_purchases = purchase_data_pd.groupby(["Gender"]).count()["Price"]
average_purchase_price =  purchase_data_pd.groupby(["Gender"]).mean()["Price"]
total_purchase_value =  purchase_data_pd.groupby(["Gender"]).sum()["Price"]
avg_per_person = total_purchase_value / total_gender

purchase_analysis_df = pd.DataFrame({"Total Purchases": total_purchases, "Average Purchase": average_purchase_price, "Total Purchase Value": total_purchase_value, "Avg Total Purchase Per Person": avg_per_person})
purchase_analysis_df["Average Purchase"] = purchase_analysis_df["Average Purchase"].map("${0:,.2f}".format)
purchase_analysis_df["Total Purchase Value"] = purchase_analysis_df["Total Purchase Value"].map("${0:,.2f}".format)
purchase_analysis_df["Avg Total Purchase Per Person"] = purchase_analysis_df["Avg Total Purchase Per Person"].map("${0:,.2f}".format)

purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 40]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
unique_players["Age Group"] = pd.cut(unique_players["Age"], bins, labels=age_labels)
age_group_total = unique_players["Age Group"].value_counts()
age_group_percent = age_group_total / total_players

age_df = pd.DataFrame({"Number of Players": age_group_total, "Percent of Players": age_group_percent})
age_df["Percent of Players"] = age_df["Percent of Players"].map("{0:,.2%}".format)
age_df = age_df.sort_index()

age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_pd["Age Group"] = pd.cut(purchase_data_pd["Age"], bins, labels=age_labels)

total_purchases_age = purchase_data_pd.groupby(["Age Group"]).count()["Price"]
average_purchase_price_age =  purchase_data_pd.groupby(["Age Group"]).mean()["Price"]
total_purchase_value_age =  purchase_data_pd.groupby(["Age Group"]).sum()["Price"]
average_per_person = total_purchase_value_age / age_group_total

age_purchase_df = pd.DataFrame({"Total Purchases": total_purchases_age, "Average Purchase Price": average_purchase_price_age, "Total Purchase Value": total_purchase_value_age, "Avg Total Purchase Per Person": average_per_person})
age_purchase_df["Average Purchase Price"] = age_purchase_df["Average Purchase Price"].map("${0:,.2f}".format)
age_purchase_df["Total Purchase Value"] = age_purchase_df["Total Purchase Value"].map("${0:,.2f}".format)
age_purchase_df["Avg Total Purchase Per Person"] = age_purchase_df["Avg Total Purchase Per Person"].map("${0:,.2f}".format)

age_purchase_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count_spenders = purchase_data_pd.groupby(["SN"]).count()["Price"]
average_spend = purchase_data_pd.groupby(["SN"]).mean()["Price"]
total_spend = purchase_data_pd.groupby(["SN"]).sum()["Price"]

top_spend_df = pd.DataFrame({"Purchase Count": purchase_count_spenders, "Average Purchase Price": average_spend, "Total Purchase Value": total_spend})
top_spend_df = top_spend_df.sort_values(["Total Purchase Value"], ascending=False).round(2)
top_spend_df["Total Purchase Value"] = top_spend_df["Total Purchase Value"].map("${0:,.2f}".format)
top_spend_df["Average Purchase Price"] = top_spend_df["Average Purchase Price"].map("${0:,.2f}".format)

top_spend_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_info_df = purchase_data_pd.loc[:, ["Item Name", "Item ID", "Price"]]
unique_item_df = item_info_df.drop_duplicates()

item_df=purchase_data_pd.groupby(['Item Name', 'Item ID']).agg({'Purchase ID': ['count'], 'Price':['sum']})

merge_item = pd.merge(item_df, unique_item_df, on="Item ID")
merge_item.columns = ['Item ID', 'Purchase Count', 'Total Purchase Value', 'Item Name', 'Item Price']
organized_item = merge_item[['Item ID', 'Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']]
organized_item = organized_item.sort_values(["Purchase Count"], ascending=False).round(2)
organized_item["Total Purchase Value"] = organized_item["Total Purchase Value"].map("${0:,.2f}".format)
organized_item["Item Price"] = organized_item["Item Price"].map("${0:,.2f}".format)

organized_item
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_info_df = purchase_data_pd.loc[:, ["Item Name", "Item ID", "Price"]]
unique_item_df = item_info_df.drop_duplicates()

item_df=purchase_data_pd.groupby(['Item Name', 'Item ID']).agg({'Purchase ID': ['count'], 'Price':['sum']})

merge_item = pd.merge(item_df, unique_item_df, on="Item ID")
merge_item.columns = ['Item ID', 'Purchase Count', 'Total Purchase Value', 'Item Name', 'Item Price']
organized_item = merge_item[['Item ID', 'Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']]
organized_item = organized_item.sort_values(["Total Purchase Value"], ascending=False).round(2)
organized_item["Total Purchase Value"] = organized_item["Total Purchase Value"].map("${0:,.2f}".format)
organized_item["Item Price"] = organized_item["Item Price"].map("${0:,.2f}".format)

organized_item
```
